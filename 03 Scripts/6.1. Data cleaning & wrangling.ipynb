```python
# Adding new columns 'weekday/weekday' and 'city' to each dataframe
# weekday/weekday: marked as 'weekday' if data is recorded on weekdays, 'weekend' if data is recored on weekend
# city: object, capitalized city names
paris_weekdays['weekday/weekend'], paris_weekdays['city'] = 'Weekday', 'Paris'
paris_weekends['weekday/weekend'], paris_weekends['city'] = 'Weekend', 'Paris'

rome_weekdays['weekday/weekend'], rome_weekdays['city'] = 'Weekday', 'Rome'
rome_weekends['weekday/weekend'], rome_weekends['city'] = 'Weekend', 'Rome'

vienna_weekdays['weekday/weekend'], vienna_weekdays['city'] = 'Weekday', 'Vienna'
vienna_weekends['weekday/weekend'], vienna_weekends['city'] = 'Weekend', 'Vienna'

ams_weekdays['weekday/weekend'], ams_weekdays['city'] = 'Weekday', 'Amsterdam'
ams_weekends['weekday/weekend'], ams_weekends['city'] = 'Weekend', 'Amsterdam'

athens_weekdays['weekday/weekend'], athens_weekdays['city'] = 'Weekday', 'Athens'
athens_weekends['weekday/weekend'], athens_weekends['city'] = 'Weekend', 'Athens'

barc_weekdays['weekday/weekend'], barc_weekdays['city'] = 'Weekday', 'Barcelona'
barc_weekends['weekday/weekend'], barc_weekends['city'] = 'Weekend', 'Barcelona'

berlin_weekdays['weekday/weekend'], berlin_weekdays['city'] = 'Weekday', 'Berlin'
berlin_weekends['weekday/weekend'], berlin_weekends['city'] = 'Weekend', 'Berlin'

buda_weekdays['weekday/weekend'], buda_weekdays['city'] = 'Weekday', 'Budapest'
buda_weekends['weekday/weekend'], buda_weekends['city'] = 'Weekend', 'Budapest'

lis_weekdays['weekday/weekend'], lis_weekdays['city'] = 'Weekday', 'Lisbon'
lis_weekends['weekday/weekend'], lis_weekends['city'] = 'Weekend', 'Lisbon'

london_weekdays['weekday/weekend'], london_weekdays['city'] = 'Weekday', 'London'
london_weekends['weekday/weekend'], london_weekends['city'] = 'Weekend', 'London'
```

```python
# Combine all data of each city
paris_all = paris_weekdays.merge(paris_weekends, how = 'outer')
rome_all = rome_weekdays.merge(rome_weekends, how = 'outer')
vienna_all = vienna_weekdays.merge(vienna_weekends, how = 'outer')
ams_all = ams_weekdays.merge(ams_weekends, how = 'outer')
athens_all = athens_weekdays.merge(athens_weekends, how = 'outer')
barc_all = barc_weekdays.merge(barc_weekends, how = 'outer')
berlin_all = berlin_weekdays.merge(berlin_weekends, how = 'outer')
buda_all = buda_weekdays.merge(buda_weekends, how = 'outer')
lis_all = lis_weekdays.merge(lis_weekends, how = 'outer')
london_all = london_weekdays.merge(london_weekends, how = 'outer')
```

```python
# Combine all cities into one dataset
airbnb = paris_all.merge(rome_all, how = 'outer'
                ).merge(vienna_all, how = 'outer'
                ).merge(ams_all, how = 'outer'
                ).merge(athens_all, how = 'outer'
                ).merge(barc_all, how = 'outer'
                ).merge(berlin_all, how = 'outer'
                ).merge(buda_all, how = 'outer'
                ).merge(lis_all, how = 'outer'
                ).merge(london_all, how = 'outer')
```

## 03. Data Cleaning

### Mixed type data

```python
# Checking mixed typed
for col in airbnb.columns.tolist():
  weird = (airbnb[[col]].map(type) != airbnb[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (airbnb[weird]) > 0:
    print (col)
```

#### There is no mixed type data column in the final dataframe

### Missing values

```python
# Checking missing values
airbnb.isnull().sum()
```

#### There is no missing value found in the dataframe

### Duplicates

## 04. Data Wrangling

```python
airbnb.info()
```

```python
# Dropping unnecessary columns
airbnb_drop = airbnb.drop(columns = ['Unnamed: 0','attr_index','rest_index','lng','b'])
```

#### I drop columns 'Unamed: 0', 'lng', 'lat' becasue it's not relevent with my analysis. For columns 'attr_index','rest_index', I already used accordingly normalized columns so I don't need these columns anymore

```python
airbnb_drop
```

```python
# Rename columns for better understanding
airbnb_drop.rename(columns = {'realSum' : 'price'}, inplace = True)
airbnb_drop.rename(columns = {'multi' : 'multi_listing'}, inplace = True)
airbnb_drop.rename(columns = {'biz' : 'business_listing'}, inplace = True)
airbnb_drop.rename(columns = {'dist' : 'city_center_dist'}, inplace = True)
```
